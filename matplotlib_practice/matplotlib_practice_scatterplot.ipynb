```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create scatter plot using plt.plot()
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create scatter plot using plt.scatter()
plt.scatter(x_axis, y_axis)
```

```python
# Skill drill 5.1.7 
# Using the matplotlib MATLAB plotting approach, make the foloowing changes to your scatter plot:
# Change the color of markers to red                         DONE
# Add a legend for the city of Chicago                       DONE
# Add a title and axes labels                                DONE
# Switch the axis so that the Fare($) data is on the x-axis  DONE
# Add a limit for the x-axis data                            DONE
# Invert the y-axis so that January is on the top            DONE

# Create the plot and change color to red
plt.scatter(y_axis, x_axis, c="red", label="Chicago")

# Add title and label axes
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Month")

# Limit x axis
plt.xlim(0, 45)

# Invert y axis so Jan is on top
plt.gca().invert_yaxis()

# Add a legend 
plt.legend()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s= [i * 3 for i in y_axis])
```

```python
# Create scatter plot using object oriented method

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Skill Drill 5.1.7 
# Using object-oriented approach, make the following changes to your bubble chart:
# Change the color of the markers to sky blue                        DONE
# Change the size of the markers to 5 times each data point          DONE
# Make the color 20% transparent                                     DONE
# Add a black edge color to the circles                              DONE
# Make the linewidth of circles 2 points                             DONE
# Add a legend for the city of Boston                                DONE
# Add a title                                                        DONE 
# Swtich the axis so that the Fare($) data is on the x-axis          DONE
# Add a limit for the x-axis data                                    DONE
# Invert the y-axis so that Jan is at the top                        DONE

# Create the plot using ax.plot(), change color, change marker size
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c="skyblue", s= [i * 5 for i in y_axis], label="Boston", alpha=0.2, edgecolor="k", linewidth=2)

# Add title and axes labels
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")

# x-axis limit
ax.set_xlim(0, 45)

# Invert y axis so Jan is on top
ax.invert_yaxis()

# Add city legend
ax.legend()
```
