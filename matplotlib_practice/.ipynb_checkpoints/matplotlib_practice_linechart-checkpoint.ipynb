```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot using MATLAB method
plt.plot(x_axis, y_axis)

# Show labels on the graph (not in lesson, tried on my own)
plt.ylabel('Total Fare ($)')
plt.xlabel('Month')
```

```python
# Create the plot with ax.plot() and fig()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with .show()
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot 
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Skill Drill 5.1.4 Using object oriented approach, make the following changes: 
# change line color to green           DONE
# change the marker to diamond         DONE
# line width to 2 points               DONE
# add a legend for the city of Boston  DONE
# add title and axes labels            DONE
# add y-axis limit                     DONE
# add grid lines                       DONE

# Create the plot
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="<", color="green", linewidth=2)

# Create label
ax.set_label("Boston")

# Create labels for the x and y axes
ax.set_xlabel('Date')
ax.set_ylabel("Fare($)")

#Set the y limit between 0 and 45
ax.set_ylim(0, 45)

# Create a title
ax.set_title("PyBer Fare by Month")

# Add the legend
ax.legend("Boston")

# Add grid lines
ax.grid()
```
