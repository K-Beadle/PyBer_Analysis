```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.pie(y_axis, labels=x_axis, autopct="%1.1f%%")
plt.show
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, labels=x_axis, explode=explode_values, autopct="%1.1f%%")
plt.show
```

```python
# Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
        labels=x_axis,
        explode=explode_values,
        colors=colors,
        autopct="%1.1f%%")
plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Skill Drill 5.1.8 
# Using the object-oriented approach, make the following changes to the pie chart:
# Add a percentage to one decimal place to each wedge of the pie                        DONE
# Increase the figure size to 8x8                                                       DONE
# Explode the two highest percentage months                                             DONE
# Add a shadow                                                                          DONE
# Add a start angle so that January is at the top                                       DONE
# Reverse the order so that the month order is in a clockwise direction                 DONE
# Add new colors of your own choosing or use the colors from the previous pie chart     DONE

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_months = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, 
       labels=x_axis,
       explode=explode_months,
       colors=colors,
       autopct="%1.1f%%",
       shadow=True,
       startangle=270,)

# Reverse order so Jan is at top
plt.gca().invert_yaxis()

plt.show()
```
